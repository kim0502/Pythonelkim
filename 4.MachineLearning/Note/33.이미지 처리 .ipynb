# Numpy를 Image로 변환하기

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Numpy Data 불러오기
import numpy as np
fruits = np.load('../Data/fruits.npy')
fruits.shape
```

```python
# Image로 저장하기
from PIL import Image
for i in range(0,300):
    image = Image.fromarray(fruits[i]).convert("L") 
    image.save('../Data/fruits/fruit_%03d.jpg'%i,'JPEG')
```

# Image를 읽어서 Numpy로 변환하기

```python
import glob
```

```python
numberOfdata = 300
imgSize = 100
```

```python
# 배열 초기화로 만들기
fruitData = np.zeros(numberOfdata*imgSize*imgSize, dtype=np.int32).reshape(numberOfdata, imgSize, imgSize)
fruitData.sum()
# 0으로 초기화 만들기(초기화하고 reshape(쪼개주기)
```

```python
i = 0
for file in sorted(glob.glob("../Data/fruits/*.jpg")):
    img = np.array(Image.open(file), dtype=np.int32)
    fruitData[i::] = img
    i += 1
```

```python
# 데이터 들어갔는지 확인 -1
fruitData.shape
```

```python
# 데이터 들어갔는지 확인 -2
fruitData.sum()
```

```python
# Image 확인
import matplotlib.pyplot as plt
plt.imshow(fruitData[0].reshape(100,100), cmap='gray_r')
plt.show()
```

```python
# Image 확인
import matplotlib.pyplot as plt
plt.imshow(fruitData[100].reshape(100,100), cmap='gray_r')
plt.show()
```
