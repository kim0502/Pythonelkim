```python
stores_1= stores

list=['한식', '용달운송', '부동산', '옷가게', '주점', '미용실', '슈퍼마켓', '분식', '세탁소', '노래방','편의점', '치킨집', '카센터', '과일채소가게', '정육점', '중국집', '당구장', '빵집', 'pc방', '카페']
stores_1.index
for i in list:
   stores_1[i] = stores_1[i][stores_1.index] / stores_1['전체상점'][stores_1.index] * 100 
```

```python
stores_1.head()
```

```python
csv_data = stores_1[['한식', '용달운송', '부동산', '옷가게', '주점', '미용실', '슈퍼마켓', '분식', '세탁소', '노래방','편의점', '치킨집', '카센터', '과일채소가게', '정육점', '중국집', '당구장', '빵집', 'pc방', '카페']]
csv_label =stores_1['자치구행정동명']
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(
    csv_data, csv_label
)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
nn_clf = KNeighborsClassifier(n_neighbors=3) #n_neighbors=3 :하이퍼 파라미터라고한다. 데이터가 많은때는 줄이고, 적을때는 올린다.
nn_clf.fit(train_data,train_label)
print("KNN train :",nn_clf.score(train_data,train_label))
print("KNN test :",nn_clf.score(test_data,test_label))
print('-' * 30)

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt_clf =DecisionTreeClassifier()
dt_clf.fit(train_data,train_label)
print("Decision Tree train :", dt_clf.score(train_data,train_label))
print("Decision Tree test :", dt_clf.score(test_data,test_label))
print('-' * 30)

# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(train_data,train_label)
print("Random Forest train :", rf_clf.score(train_data,train_label))
print("Random Forest test :", rf_clf.score(test_data,test_label))
print('-' * 30)

# AdaBoost
from sklearn.ensemble import AdaBoostClassifier
ad_clf =AdaBoostClassifier()
ad_clf.fit(train_data,train_label)
print("AdaBoost train :", ad_clf.score(train_data,train_label))
print("AdaBoost test :", ad_clf.score(test_data,test_label))
print('-' * 30)

# Neural Net
from sklearn.neural_network import MLPClassifier
net_clf = MLPClassifier()
net_clf.fit(train_data,train_label)
print("Neural Net train :", net_clf.score(train_data,train_label))
print("Neural Net test :", net_clf.score(test_data,test_label))
print('-' * 30)

# Naive Bayes 
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(train_data,train_label)
print("Naive Bayes train :", nb_clf.score(train_data,train_label))
print("Naive Bayes test :", nb_clf.score(test_data,test_label))
print('-' * 30)

# # QDA
# from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
# qda_clf = QuadraticDiscriminantAnalysis()
# qda_clf.fit(train_data,train_label)
# print("QDA train :", qda_clf.score(train_data,train_label))
# print("QDA test :", qda_clf.score(test_data,test_label))
# print('-' * 30)
```
