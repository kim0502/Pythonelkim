## Data 정규화

```python
# weight, height
csv['weight'] = csv['weight'] / csv['weight'].max()
csv['height'] = csv['height'] / csv['height'].max()
```

```python
csv.head()
```

```python
# label
bclass ={'thin':[1,0,0], 'normal':[0,1,0], 'fat':[0,0,1]}
y = []
for i,v in enumerate(csv['label']):
    y.append(bclass[v])

y
```

```python
# Training
X_train = csv[['height','weight']][0:15000]
y_train = y[0:15000]
```

```python
# Test
X_test = csv[['height','weight']][15001:20001]
y_test = y[15001:20001]
```

# Deep Learning 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(512, activation ='relu', input_shape=(2,)))
model.add(keras.layers.Dense(3, activation='softmax'))
```

```python
from pickletools import optimize # 경사하강법

from sklearn import metrics 


model.compile(
    optimizer = 'rmsprop',
    loss = 'categorical_crossentropy', # 실수
    metrics=['accuracy']
)
```

```python
# 데이터 훈련하기
import numpy as np

model.fit(
    np.array(X_train),
    np.array(y_train),
    epochs=50
)
```

---
# 테스트 데이터로 평가하기

```python
score = model.evaluate(X_test, np.array(y_test))
print('loss :', score[0])
print('accuracy :', score[1])
```
