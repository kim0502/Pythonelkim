```python
# traget Data
y = data['admit']
# y와 같이 출력층은 1개이다.
y.head()
```

## Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
# 1번째 Hidden Layer : 256 노드 (2의 배수, 임의의 갯수(Hyper Parameter))
# activation = sigmoid, tanh, relu, softmax, leakyRelu
model.add(keras.layers.Dense(256, activation = 'tanh', input_shape=(3,))) # 은닉층
model.add(keras.layers.Dense(1, activation='sigmoid')) # 출력층
```

```python
# 손실함수
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy') 
# optimizer= 정답값(y) - 에측값(y^)
# optimizer : learning rate를 알맞게 조정해줌
# - adam
# - adagrad
# - asadelta
# - rmsprop
# - sgd
# - less : 손실함수
# - binary_crossentropy : 분류 및 혹률 예측
# - metrics : 모델의 평가 기준
```

```python
# 학습 시키기
import numpy as np
model.fit(np.array(x), np.array(y), epochs=500)
```

```python
# 예측하기

model.predict(
    [
        [700,3.7,1],
        [700,3.7,2],
        [700,3.7,3],
        [700,3.7,4]
    ]
)
```
