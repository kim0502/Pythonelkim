### dataFrame을 Database로 Insert하기

```python
import pandas as pd 
from sqlalchemy import create_engine
column = ['rank','title']
movies = pd.read_csv("../Data/ridi_page1.csv")
movies.columns =column
movies.head()
```

```python
engine = create_engine("mysql+pymysql://root:qwer1234@127.0.0.1:3306/Books")
conn = engine.connect()
```

```python
# - 데이터 프레임 movies
# - sql table movie
# - sql 스키마 daum
```

```python
movies.to_sql(name='new_table', con=engine, if_exists='append', index=False)
```

```python
daum = pd.read_sql_table('new_table', conn)
daum
```
