```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('http://www.baskinrobbins.co.kr/')
```

```python
driver.save_screenshot("../Data/002.png") # 화면을 관리하고 있는지 여부를 확인하기 위해 수행
```

```python
gu_list_raw = driver.find_element(By.XPATH, '//*[@id="nform"]/fieldset/p[1]/span[1]/select')
gu_list =gu_list_raw.find_elements(By.TAG_NAME,'option')

gu_names= [option.get_attribute('value') for option in gu_list]
gu_names.remove('')
gu_names
```

```python
from bs4 import BeautifulSoup

page_source = driver.page_source 
soup= BeautifulSoup(page_source, "html.parser")
print(soup)
```

```python
lists=soup.select("#content > div > div.store_search > div.list > div > ul > li > article > a > address > span")
# lists
stores = soup.select("#content > div > div.store_search > div.list > div > ul > li > article > h3 > a ")
stores
```

```python
for store in stores:
    print(store.string)
```

```python
for list in lists:
    print(list.string)
```

# 데이터 수집

```python
# 화면에 그림을 그리는 시간이 필요하므로 이를 delay시키자!
storeList =[]
addressList=[]
import time
for gu in gu_names:
    element = driver.find_element(By.XPATH, '//*[@id="nform"]/fieldset/p[1]/span[1]/select')
    element.send_keys(gu)

    #time.sleep(2)
    
    xpath = '//*[@id="nform"]/fieldset/div[2]/button'
    element_sle_gu = driver.find_element(By.XPATH, xpath).click()

    #time.sleep(1)
    page_source = driver.page_source 
    soup= BeautifulSoup(page_source, "html.parser")
    
    lists=soup.select("#content > div > div.store_search > div.list > div > ul > li > article > a > address > span")
    for list in lists:
        addressList.append(list.string)
    stores = soup.select("#content > div > div.store_search > div.list > div > ul > li > article > h3 > a ")
    for store in stores:
        storeList.append(store.string)
   
    
```

```python
#print(addressList)
import pandas as pd
```

```python
df = pd.DataFrame(data=storeList,columns=['매장'])
df
```

```python
df['주소'] = pd.Series(addressList)
df
```

```python
df.to_csv("../Data/Baskin.csv",encoding='utf-8', index=False)
```
