```python
# Library
from bs4 import BeautifulSoup
import urllib.request as req

# Site Address
url = "https://news.daum.net/ranking"
html =req.urlopen(url)
soup = BeautifulSoup(html, "html.parser")
print(soup)
```

```python
news=soup.select("div.box_g.box_news_issue > ul > li > div > div >strong > a")
newsList = []
for i in news:
   #print(i.string)
   newsList.append(i.string.strip())
newsList
```

```python
import pandas as pd
from sqlalchemy import create_engine
                                
```

```python
newsDataFrame = pd.DataFrame(newsList, columns=['mainTitle'])
newsDataFrame.head()
```

```python
newsDataFrame.to_csv("../Data/newDataFrame.csv",encoding='utf-8', index=False)
```
