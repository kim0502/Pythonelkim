# 데이터의 정규화(Normallization)
- 강도와 살인은 2자릿수 이나 절도와 폭력은 4자리입니다.
- 각각의 범죄 발생건수를 비숫한범위에 놓고 비교하는 것이 편리합니다.
- 각 항목의 값을 최대1로 계산하여 그 비중 자체로 분석하는 것이 좋습니다.

```python
# 머신러닝 모듈 사용하기
# sklearn의 preprocessing기능을 사용하면 최대값과 최소값을 비교하면서 정규화가 진행
from sklearn import preprocessing
```

```python
# 정규화 진행할 컬럼 설정
col = ['강간','강도','살인','절도','폭력']
```

```python
# 정규화를 진행할 컬럼의 데이터 설정
x = crime_anal[col].values
x
# x.dtype 확인을 해보면 int이다. 
# 정규화(0과1사이의 값)를 시킬려면 float(소수)로 바꾸주어야한다.
```

```python
# 최대값과 최소값의 값 가져오기
min_max_scaler = preprocessing.MinMaxScaler()
min_max_scaler
```

```python
# 정규화(0과1사이의 값) 시킬 최종 값은 비율로 계산되기 때문에 float설정
x_scaled = min_max_scaler.fit_transform(x.astype(float))
x_scaled
# min_max_scaler에 x를 넣어서 함수에 적용 하는것을 fitting이라고 한다.
```

```python
# 정규화된 범죄건수가 있는 새로운 데이터프레임 생성
crim_anal_norm = pd.DataFrame(x_scaled, columns=col, index=crime_anal.index)
crim_anal_norm
```

```python
col2 = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crim_anal_norm[col2] = crime_anal[col2]
crim_anal_norm.head()
```

### heatmap을 이용한 전체 검거율 확인

```python
# 검거 최대값 가져오기
tmp_max = crim_anal_norm['검거'].max()
# 검거의 최대값을 100으로 설정하고 검거율 계산하기
crim_anal_norm['검거'] =crim_anal_norm['검거'] / 100

# 검거율을 기준으로 정렬하기
crim_anal_norm_sort = crim_anal_norm.sort_values(by='검거', ascending=False)
crim_anal_norm_sort.head()
```
