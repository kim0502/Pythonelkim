```python
import pandas as pd
import urllib.request
import folium
import json
```

```python
pizza = pd.read_csv("../Data/pizzaschool_map.csv")
pizza
```

```python
domino= pd.read_csv("../Data/domino3조.csv")
domino
```

```python
map = folium.Map(
    location=[37.5502,126.982],
    zoom_start=11
)
for n in pizza.index:
    folium.Marker([pizza['위도'][n], pizza['경도'][n]],
                   popup=pizza['가맹점명'][n]
    ).add_to(map)
    folium.CircleMarker([pizza['위도'][n],
                         pizza['경도'][n]],
                         popup=pizza['주소'][n],
                         radius=5,
                         fill_color='skyblue',
                         fill_opacity=0.3,
    ).add_to(map)

for n in domino.index:
    folium.Marker([domino['lat'][n], domino['lng'][n]],
                   popup = domino['지점명'][n]
    ).add_to(map)
    folium.Marker([domino['lat'][n],
                         domino['lng'][n]],
                         popup = domino['매장주소'][n],
                         color = 'red',
                         radius = 5,
                         fill_color ='red',
                         fill_opacity = 0.3,
                         icon = folium.Icon(color='red', icon="star")
                         
    ).add_to(map)
    
map
```
