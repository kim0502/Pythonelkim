```python
train3.loc[train3['Fare'] < 15, 'FareType'] = 'Low'
train3.loc[(train3['Fare'] >= 15)& (train['Fare']<=30), 'FareType'] = 'Medium'
train3.loc[train3['Fare'] > 30, 'FareType'] = 'High'
train3
```

```python
def check_fare_type(fare):
        if fare < 15:
            return 'Low'
        elif fare >=15 and fare < 30:
            return "Medium"
        else:
            return "Hight"
```

```python
train3['FareType']= train3['Fare'].apply(check_fare_type)
train3[['Fare','FareType']].head(10)
```
