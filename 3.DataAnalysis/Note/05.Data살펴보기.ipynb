### 자동차 연비 Dataset
- mpg : 연비
- cylinders : 실린더 수
- displacement : 배기량
- horsepower : 출력
- weight : 차중
- acceleration : 가속 능력
- model_year : 출시 년도
- origin : 제조국(1:USA, 2:EU, 3:JPN)
- name : 모델명

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration',' model_year','origin','name']
df.head()
```

### Pandas의 내장 그래프도구

```python
df =pd.read_excel('../Data/남북한발전전력량.xlsx')
df
```

```python
df_sn = df.iloc[[0,5], 2:]
df_sn.index = ['South','North']
df_sn
```

```python
df_sn = df.loc[[0,5], '1990':]
df_sn.index = ['South','North']
df_sn
```

```python
df_sn = df.loc[[0,5], '1990':]
df_sn.index = ['South','North']
df_sn
df_sn.columns =df_sn.columns.map(int) # 열 이름의 자료형을 정수형으로 변형, x,y차트그릴때 중요!!!
```

```python
df_sn = df_sn.loc[:, '1990':]
df_sn
```

```python
# 선 그래프 그리기
df_sn.plot()
```

```python
df_sn = df.loc[[0,5], '1990':'2015']
df_sn.index = ['South','North']
df_sn
df_sn.columns =df_sn.columns.map(int)
```

```python
df_sn = df_sn.loc[:, '1990':'2015']
df_sn
```

```python
df2=df_sn.T
```

```python
df2.plot()
```

```python
# 막대 그래프 그리기
df2.plot(kind='bar')
```

```python
# 히스토그램
df2.plot(kind ='hist')
```

```python
df = pd.read_csv("../Data/auto-mpg.csv", header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration',' model_year','origin','name']
df.head()
```
