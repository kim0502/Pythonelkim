### CCTV 현황

```python
# CCTV에서 소계로 오름차순 정렬
CCTV_Seoul.sort_values(by='소계', ascending=True).head()
```

---
#### 최근 3년간 CCTV증가율 구해서 CCTV가 제일 많이 증가한 구를 검색후 정렬


```python
a1 = CCTV_Seoul.iloc[:,2:7].sum(axis=1)
a2 = CCTV_Seoul.iloc[:,7:10].sum(axis=1)
```

```python
CCTV_Seoul['최근증가율'] =  (a2 / a1) * 100
```

```python
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

> 결과를 보면 최근 3년간의 CCTV가 2015년 이전에 비해 가장 많이 증가한 곳은 중구, 영등포구, 성동구, 금천구, 광진구 순이다.

# CCTV와 인구현황 합치기

```python
Data_result = pd.merge(CCTV_Seoul, pop_Seoul, on= '구별')
Data_result.head()
```

```python
Data_result.set_index('구별', inplace=True)
Data_result.head()
```

```python
# 구별을 index로 만들기
delColumn = Data_result.columns[2:9]

for i in delColumn:
    del Data_result[i]
Data_result.head()
```

```python
import numpy as np
```

```python
# 고령자 비율과 소계의 상관계수
np.corrcoef(Data_result['고령자비율'], Data_result['소계']) 
# 결과 : 역상관관계가 나온다.
```

```python
# 고령자와 소계의 상관계수
np.corrcoef(Data_result['고령자'], Data_result['소계'])
# 결과 : 상관이 없다.
```

```python
# 외국인 비율과 소계의 상관계수 
np.corrcoef(Data_result['외국인비율'], Data_result['소계'])
# 결과 :상관이 없다.
```

```python
# 외국인과 소계의 상관계수
np.corrcoef(Data_result['외국인'], Data_result['소계'])
# 결과 : 상관없다.
```

```python
# 인구수와 소계의 상관계수
np.corrcoef(Data_result['인구수'], Data_result['소계'])
# 결과 : 양의 상관이 있다.
```

> 고령자 비율의 상관계수를 보면 역상관이 발생하고 이는고령자가 아닌젊은 층 비율이 높은 곳에 CCTV가 많이 설치되는 경행을 보이며
> 주로 인구수와 양의 상관을 보이고 있어 CCTV 댓수는 인구수에 비례하여 증가하는 경향을 보인다.
